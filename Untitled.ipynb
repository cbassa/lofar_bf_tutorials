```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
nsub = 10
isub = 9
nsamp = 11700000
s0 = np.fromfile("L645495_SAP000_B000_S0_P000_bf.raw", dtype="float32", count=nsub*nsamp)[isub::nsub]
s1 = np.fromfile("L645495_SAP000_B000_S1_P000_bf.raw", dtype="float32", count=nsub*nsamp)[isub::nsub]
```

```python
plt.figure(figsize=(20,10))
plt.plot(s0[:1000])
plt.plot(s1[:1000])
```

```python
cin = s0+1j*s1
```

```python
nchan = 1000
cout = np.fft.fft(cin.reshape(-1, nchan))
t = np.arange(nsamp)*5.12e-6
f = np.fft.fftshift(np.fft.fftfreq(nsamp, d=5.12e-6))
amp = 10.0*np.log10(np.absolute(np.fft.fftshift(cout, axes=1)))
vmin = np.mean(amp)-2.0*np.std(amp)
vmax = np.mean(amp)+6.0*np.std(amp)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(amp.T, aspect='auto', origin='lower', vmin=vmin, vmax=vmax, extent=[t[0], t[-1], f[0], f[-1]])
plt.grid()
plt.colorbar()
```
